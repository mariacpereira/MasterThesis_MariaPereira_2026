#### Topic missing: is it a Governance Indicator?

```python
esg_series["Topic"].isna().sum()
```

```python
esg_series[esg_series["Topic"].isna()][["Series Code", "Indicator Name"]]
```

ER.H2O.FWST.ZS – “Level of water stress: freshwater withdrawal as a % of available freshwater resources”
is indeed an environmental indicator (not governance-related).

```python
# Correct the "Topic" issue 
esg_series.loc[esg_series["Series Code"]=="ER.H2O.FWST.ZS",
               "Topic"] = "Environment: Water Resources"
```

```python
esg_series["Topic"].isna().sum()
```

```python
# Filter Governance indicators
gov_series = esg_series[esg_series["Topic"].str.startswith("Governance", na=False)].copy()

print("Number of Governance Indicators:", gov_series["Series Code"].nunique())
display(gov_series[["Series Code", "Indicator Name", "Topic"]].sort_values("Indicator Name"))
```

Check this indicator: IC.LGL.CRED.XQ	Strength of legal rights index (0=weak to 12=s...	Governance: Human Rights

```python
code_to_check = "IC.LGL.CRED.XQ"
subset = esg_data[esg_data["Indicator Code"] == code_to_check]
```

```python
print(f"Number of indicator lines {code_to_check}: {subset.shape[0]}")
display(subset.head())
```

During metadata validation, this indicator appeared under “Governance: Human Rights” in the ESG dataset.
However, this classification is inconsistent: it belongs to the World Development Indicators and measures credit market legal protection, not governance or human rights.
Therefore, it was excluded from the Governance dataset to maintain conceptual consistency.

```python
# Exclude the indicator 
gov_series = gov_series[gov_series["Series Code"] != "IC.LGL.CRED.XQ"].copy()
```

```python
print("Number of Governance Indicators:",
      gov_series["Series Code"].nunique())
```

#### Is there any EU27 Country missing from the dataset?

```python
present = sorted(set(esg_data["Country Name"]) & set(EU27_NAMES))
missing = [c for c in EU27_NAMES if c not in present]

print(f"EU-27 countries present ({len(present)}/{len(EU27_NAMES)}):")
print(present)
print()
print(f"Missing ({len(missing)}):")
print(missing)
```

```python
EU27_ISO3 = [
    "AUT","BEL","BGR","HRV","CYP","CZE","DNK","EST","FIN","FRA",
    "DEU","GRC","HUN","IRL","ITA","LVA","LTU","LUX","MLT","NLD",
    "POL","PRT","ROU","SVK","SVN","ESP","SWE"
]

# Only Governance
esg_gov = esg_data[esg_data["Indicator Code"].isin(esg_data["Indicator Code"])]

# Indicator per country
coverage = (esg_gov.groupby("Country Code")["Indicator Code"].nunique().reset_index(name="num_indicators"))
coverage["expected"] = len(esg_data)
coverage["missing"] = coverage["expected"] - coverage["num_indicators"]

print(coverage.sort_values("missing", ascending=False))
```

The results show that all countries have at least some Governance indicators available (num_indicators > 0), confirming that there is data coverage for all countries in the ESG dataset.
